```python
import pandas as pd
import numpy as np
import urllib.request
import requests
```

```python


# Google Maps Geocoding API and Distance Matrix API endpoints
geocoding_api_url = 'https://maps.googleapis.com/maps/api/geocode/json'
distance_matrix_api_url = 'https://maps.googleapis.com/maps/api/distancematrix/json'

# Google API key
api_key = 'your key'

# Read destinations from the file
with open('dests.txt', 'r') as file:
    destinations = [line.strip() for line in file.readlines()]

# Initialize an empty dataframe
df = pd.DataFrame(columns=['Target', 'Distance_km', 'Duration', 'Longitude', 'Latitude'])

# Process each destination
for destination in destinations:
    # Get the latitude and longitude of the destination using Geocoding API
    geocoding_params = {
        'address': destination,
        'key': api_key
    }
    geocoding_response = requests.get(geocoding_api_url, params=geocoding_params)
    geocoding_data = geocoding_response.json()
    
    if geocoding_data['status'] == 'OK':
        location = geocoding_data['results'][0]['geometry']['location']
        latitude = location['lat']
        longitude = location['lng']

        # Get the distance and duration using Distance Matrix API
        distance_matrix_params = {
            'origins': 'Tel Aviv',
            'destinations': destination,
            'key': api_key
        }
        distance_matrix_response = requests.get(distance_matrix_api_url, params=distance_matrix_params)
        distance_matrix_data = distance_matrix_response.json()

        if distance_matrix_data['status'] == 'OK':
            distance = distance_matrix_data['rows'][0]['elements'][0]['distance']['text']
            duration = distance_matrix_data['rows'][0]['elements'][0]['duration']['text']

            # Append the data to the dataframe
            df = df.append({
                'Target': destination,
                'Distance_km': distance,
                'Duration': duration,
                'Longitude': longitude,
                'Latitude': latitude
            }, ignore_index=True)
        else:
            print(f"Distance Matrix API Error: {distance_matrix_data['status']}")
    else:
        print(f"Geocoding API Error: {geocoding_data['status']}")

# Display the dataframe
df
```

```python
# Remove commas from 'Distance_km' column
df['Distance_km'] = df['Distance_km'].str.replace(',', '')
# Convert 'Distance_km' column to numeric dtype
df['Distance_km'] = df['Distance_km'].str.replace(' km', '')

# Convert 'Distance_km' column to numeric dtype
df['Distance_km'] = df['Distance_km'].astype(float)

# Sort the dataframe by 'Distance_km' column in descending order
sorted_df = df.sort_values('Distance_km', ascending=False)

# Get the three furthest cities from Tel Aviv
furthest_cities = sorted_df.head(3)

furthest_cities
```
